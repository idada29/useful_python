```python
import numpy as np
import pandas as pd
import folium as fo
%matplotlib inline
```

```python
# A custom pane is created to contain map elements, stored in a variable 'map'
map = fo.Map(z_index = 625)
map
```

```python
# Import csv with locations, latitude and longitude
locations = pd.read_csv('Location.csv')
locations.head(5)
```

```python
# Converting columns of latitude and longitude to ease iterations
location_latitude = list(locations['Lattitude'])
location_longitude = list(locations['Longitude'])
location_name = list(locations['Name '])
location_countries = list(locations['Countries'])
```

```python
# Define a function to specify icon colours

def icon_maker(name):
    global location_countries
    if (name == 'USA'):
        return 'blue'
    elif (name == 'UK'):
        return 'red'
    else:
        return 'green'
```

```python
# Create a 'child' with subsequent paramters for introducing specific locations
locations = fo.FeatureGroup(name='My Map')

# The zip was added to prevent unpacking error issues
for lat,lon,name,countries in zip(location_latitude,location_longitude,location_name,location_countries):
    locations.add_child(fo.Marker(location=[lat,lon],popup=name,icon=fo.Icon(color=icon_maker(countries))))
```

```python
map.add_child(locations)
```
